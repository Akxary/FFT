```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
columns = ['N', 'l', 'I', 'Nan' ]
df = pd.read_csv('D:\\Учёба\\нтц уп ран\\C60F\\raman_to_process\\3 Gpa_405_50s_2sum_02mW.csv',
                 delimiter=',', header=None, skiprows=1, names=columns, decimal=',')
df1 = df[['l', 'I']]
df1 = df1.loc[df1['l'] > 500]
```

```python
plt.plot(df1['l'], df1['I'])
n = 5
df2 = df1.query('l<1000 or l>2000')
cof = np.polyfit(df2['l'],df2['I'], n)
plt.plot(df1['l'], [sum([math.pow(x, n - i) * cof[i] for i in range(n+1)]) for x in df1['l']])
```

```python
df3 = pd.DataFrame()
df3['I'] = df1['I'] - [sum([math.pow(x, n - i) * cof[i] for i in range(n+1)]) for x in df1['l']]
df3['l'] = df1['l']
df3['dI'] = [None] + [(df3.iloc[i+1]['I'] - df3.iloc[i]['I']) / (df3.iloc[i+1]['l'] - df3.iloc[i]['l'])
                      for i in range((df3.shape[0]) - 1)]
df3['d2I'] = [None, None] + [(df3.iloc[i+1]['dI'] - df3.iloc[i]['dI']) / (df3.iloc[i+1]['l'] - df3.iloc[i]['l'])
                      for i in range(1, (df3.shape[0]) - 1)]
df3['d3I'] = [None, None, None] + [(df3.iloc[i+1]['d2I'] - df3.iloc[i]['d2I']) / (df3.iloc[i+1]['l'] - df3.iloc[i]['l'])
                      for i in range(2, (df3.shape[0]) - 1)]
df3['d4I'] = [None, None, None, None] + [(df3.iloc[i+1]['d3I'] - df3.iloc[i]['d3I']) / (df3.iloc[i+1]['l'] - df3.iloc[i]['l'])
                      for i in range(3, (df3.shape[0]) - 1)]
fig, axe = plt.subplots(5,1, figsize=(12, 12))
axe[0].plot(df3['l'], df3['I'])
axe[1].plot(df3['l'], df3['dI'], '-g')
axe[2].plot(df3['l'], df3['d2I'], '-r')
axe[3].plot(df3['l'], df3['d3I'], '-y')
axe[4].plot(df3['l'], df3['d4I'], '-b')
plt.tight_layout()
```
