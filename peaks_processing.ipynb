```python
import scipy as sc
import matplotlib.pyplot as plt
import math
import pandas as pd
import numpy as np
columns = ['N', 'l', 'I', 'Nan' ]
df = pd.read_csv('D:\\Учёба\\нтц уп ран\\C60F\\raman_to_process\\3 Gpa_405_50s_2sum_02mW.csv',
                 delimiter=',', header=None, skiprows=1, names=columns, decimal=',')
df1 = df[['l', 'I']]
df1 = df1.loc[df1['l'] > 500]
```

```python
val1 = df1['I'].to_numpy()
vp = sc.signal.detrend(val1)
vp1, heights = sc.signal.find_peaks(vp, width=5, height = 200)
print(vp1)
print(heights)
plt.plot(vp, '-')
plt.plot(vp1, vp[vp1], 'x')
```
